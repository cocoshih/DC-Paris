```python
from PIL import Image, ImageDraw, ImageFont

def fit_text_to_box(text, max_width, font_path="arial.ttf"):

    font_size = 50  
    while True:
        font = ImageFont.truetype(font_path, font_size)
        left, _, right, _ = font.getbbox(text) 
        text_width = right - left  
        if text_width <= max_width:
            break  
        font_size -= 1  
    return font_size
text_to_display = "This is a long piece of text that needs to fit within a box."
optimal_font_size = fit_text_to_box(text_to_display, 300) 
print(f"Optimal font size for text '{text_to_display}' is: {optimal_font_size}")
```

```python
max_width=1000

def wrap_text(text, max_width):
    words = text.split()
    wrapped_text = []
    current_line = ""
    for word in words:
        if width(current_line + word) <= max_width:
            current_line += word + " "
        else:
            wrapped_text.append(current_line.strip())
            current_line = word + " "
    wrapped_text.append(current_line.strip())
    return "\n".join(wrapped_text)


long_text = "This is a very long sentence that needs to be wrapped to fit within a specific width. Coco is the best & Paul is second best, and this hackathon is really great because we are going to win all prize"
wrapped_result = wrap_text(long_text, max_width) 
print(wrapped_result) 
 
```
